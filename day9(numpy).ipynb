```python
import numpy as np
np.random.seed(42)
hours_studied=np.random.randint(1,10, size=50)
attendance=np.random.randint(50,100, size=50)
previous=np.random.randint(40,95,size=50)
predicted=(hours_studied*5)+(attendance*0.3)+(previous*0.5)
dataset=np.column_stack([hours_studied, attendance, previous, predicted])
print(dataset)
```

```python
import numpy as np
rng=np.random.default_rng(seed=42)
hours_studied=rng.integers(1,10, size=50)
attendance=rng.integers(50,100, size=50)
previous=rng.integers(40,95,size=50)
predicted=(hours_studied*5)+(attendance*0.3)+(previous*0.5)
dataset=np.column_stack([hours_studied, attendance, previous, predicted])
print(dataset)
```
