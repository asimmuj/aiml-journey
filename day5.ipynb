```python
def grade_calculator(avg):
    if avg>=90:
        print("A+")
    elif avg>=80:
        print("A")
    elif avg>=65:
        print("B")
    elif avg>=50:
        print("C")
    elif avg>=35:
        print("D")
    else:
        print("F")

marks=[99,98,97,96,95]
avg= sum(marks)/len(marks)
print(marks)
grade_calculator(avg)
```

```python
def add_student(records, name, marks):
    average=sum(marks)/len(marks)
    records[name]={
    "name": name,
    "marks": marks,
    "average": average}
    
def show_records(records):
    for name, data in records.items():
        print(f"{name}: avg = {data['average']}")

students={}
add_student(students, "zeeshan", [10,10,10,10])
add_student(students, "mujtaba", [100,100,100,100])
show_records(students)
```
