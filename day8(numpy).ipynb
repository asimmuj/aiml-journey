```python
import numpy as np

marks = np.array([72, 85, 91, 60, 78, 88, 95, 55, 70, 83])

print(f"Class average : {np.mean(marks)}")
print(f"Highest mark  : {np.max(marks)}")
print(f"Lowest mark   : {np.min(marks)}")
print(f"Std deviation : {np.std(marks)}")
print(f"Students above 75: {np.sum(marks > 75)}")
print(f"Their marks   : {marks[marks > 75]}")
```
