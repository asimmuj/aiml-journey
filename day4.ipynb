```python
for i in range(1,21):
    if i%3==0 and i%5==0:
        print("fizzbuzz")
    elif i%3==0:
        print("fizz")
    elif i%5==0:
        print("buzz")
    else:
        print(i)
```

```python
marks=[99, 98, 92, 91,72, 93]
marks.append(20)
marks.remove(72)
avg= sum(marks)/len(marks)
print(marks[1:3])
print(marks)
if avg>=90:
    print("A+")
elif avg>=80:
    print("A")
elif avg>=65:
    print("B")
elif avg>=50:
    print("C")
elif avg>=35:
    print("D")
else:
    print("F")
```
